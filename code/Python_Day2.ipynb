```python
data = pd.read_csv("../data/gapminder_gdp_americas.csv", index_col = "country")
```

```python
df = data.drop('continent', axis = 1)
df.head(3)
```

```python
years = df.columns.str.strip('gdpPercap_')
```

```python
df.columns = years.astype(int)
```

```python
gdp_mexico = df.loc['Mexico']
```

```python
gdp_canada = df.loc['Canada']
```

```python
fig = plt.gcf()
plt.plot(years, gdp_mexico, 'r-',label = "Mexico")
plt.plot(years, gdp_canada, 'b--',label = "Canada")
plt.legend(loc = 'upper left')
plt.ylabel ('GDP per cap')
plt.xlabel('Year')
fig.savefig('../data/NA_gdp.png')
```

```python
def print_date(year, month, day):
    joined = str(year) + '-' + str(month) + '-' + str(day)
    print(joined)
```

```python
print_date(2018,'08', 16)
```

```python
print_date(16, '08', 2018)
```

```python
print_date(day = 16, month = '08', year = 2018)
```

```python
def average(values):
    if len(values) == 0:
        return(None)
    return (sum(values / len(values)))
```

```python
print(average([ 1, 3, 5, 7, 2, 5]))
```
